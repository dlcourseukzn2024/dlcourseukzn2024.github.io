### Creating a Tuple in Python
In Python, tuples are created by placing a sequence of values separated by a comma with or without the use of parentheses for grouping the data sequence. Tuples can contain any number of elements and of any datatype (like strings, integers, lists, etc.).

```python
Tuple1 = ()
print("Initial empty Tuple: ")
print(Tuple1)
Tuple1 = ('Geeks', 'For', 'Africa')
print("\nTuple with the use of String: ")
print(Tuple1)
list1 = [1, 2, 4, 5, 6]
print("\nTuple using List: ")
print(tuple(list1))
Tuple1 = tuple('Geeks')
print("\nTuple with the use of function: ")
print(Tuple1)
Tuple1 = (0, 1, 2, 3)
Tuple2 = ('python', 'google')
Tuple3 = (Tuple1, Tuple2)
print("\nTuple with nested tuples: ")
print(Tuple3)
```

## Set Data Type in Python
In Python, a Set is an unordered collection of data types that is iterable, mutable, and has no duplicate elements. The order of elements in a set is undefined though it may consist of various elements.

Sets can be created by using the built-in `set()` function with an iterable object or a sequence by placing the sequence inside curly braces, separated by a comma. The type of elements in a set need not be the same, various mixed-up data type values can also be passed to the set.

```python
set1 = set()
print("Initial blank Set: ")
print(set1)
set1 = set("DSI-Africa")
print("\nSet with the use of String: ")
print(set1)
set1 = set(["Hello", "To", "Afica"])
print("\nSet with the use of List: ")
print(set1)
set1 = set([1, 2, 'South', 4, 'For', 6, 'South'])
print("\nSet with the use of Mixed Values")
print(set1)
```

### Access Set Items
Set items cannot be accessed by referring to an index, since sets are unordered the items have no index. But you can loop through the set items using a for loop, or ask if a specified value is present in a set, by using the in the keyword.

```python
set1 = set(["Hello", "To", "Africans"])
print("\nInitial set")
print(set1)
print("\nElements of set: ")
for i in set1:
	print(i, end=" ")
print("Hello" in set1)
```

## Dictionary Data Type in Python
A dictionary in Python is an unordered collection of data values, used to store data values like a map, unlike other Data Types that hold only a single value as an element, a Dictionary holds a key: value pair. Key-value is provided in the dictionary to make it more optimized. Each key-value pair in a Dictionary is separated by a `colon :` , whereas each key is separated by a `comma`.

In Python, a Dictionary can be created by placing a sequence of elements within curly `{}` braces, separated by `comma`. Values in a dictionary can be of any datatype and can be duplicated, whereas keys cant be repeated and must be immutable. The dictionary can also be created by the built-in function `dict()`. An empty dictionary can be created by just placing it in curly braces`{}`.

Note  Dictionary keys are case sensitive, the same name but different cases of Key will be treated distinctly.

```python
Dict = {}
print("Empty Dictionary: ")
print(Dict)
Dict = {1: 'South', 2: 'For', 3: 'Africa'}
print("\nDictionary with the use of Integer Keys: ")
print(Dict)
Dict = {'Name': 'Africa', 1: [1, 2, 3, 4]}
print("\nDictionary with the use of Mixed Keys: ")
print(Dict)
Dict = dict({1: 'DSI', 2: 'For', 3: 'South'})
print("\nDictionary with the use of dict(): ")
print(Dict)
Dict = dict([(1, 'South'), (2, 'For')])
print("\nDictionary with each item as a pair: ")
print(Dict)
```

### Accessing Key-value in Dictionary
In order to access the items of a dictionary refer to its key name. Key can be used inside square brackets. There is also a method called `get()` that will also help in accessing the element from a dictionary.

```python
Dict = {1: 'DSI-A', 'name': 'For', 3: 'Africa'}
print("Accessing a element using key:")
print(Dict['name'])
print("Accessing a element using get:")
print(Dict.get(3))
```

# Functions
Programming basically entails the design and articulation of a sequence of instructions to solve a given problem.

**Functions:** are self-contained blocks of code that perform a specific function, hence the name. Every function has an input. That input is transformed in some way, and is then typically spit back out. Functions can be user-defined, but Python has several built in ones. In fact, we have already been using the `print()` function to display some of the data types already discussed. There is really no limit to what a function can do.   

Functions are an abstraction of a sequence of instructions, that is written once but can be called/used multiple times.

```python
import math
def prime(v): #the name of the function is prime
    for i in range(2,int(math.sqrt(v)+1)): #repeat each number in the sequence 2, 3, ... sqrt(v)+1.
        if v%i == 0:  #do the following instruction only if this expression is True.
            return False
    return True

print(prime(4)) #calling the prime function with 4 as a parameter.
print(prime(7))
```

## NumPy Arrays
NumPy arrays are unique in that they are more flexible than normal Python lists. They are called *ndarrays* since they can have any number $(n)$ of dimensions $(d)$. They hold a collection of items of any one data type and can be either a vector (one-dimensional) or a matrix (multi-dimensional). NumPy arrays allow for fast element access and efficient data manipulation.

```python
list1 = [1,2,3,4]
```

To convert this to a one-dimensional `ndarray` with one row and four columns, we can use the `np.array()` function:

```python
array1 = np.array(list1)
print(array1)
```

To get a two-dimensional `ndarray` from a list, we must start with a Python list of lists:

```python
list2 = [[1,2,3],[4,5,6]]
array2 = np.array(list2)
print(array2)
```

Many operations can be performed on NumPy arrays which makes them very helpful for manipulating data:

* Selecting array elements

* Slicing arrays

* Reshaping arrays

* Splitting arrays

* Combining arrays

* Numerical operations (min, max, mean, etc)

```python
# Creating an array of numbers of a specified range
arr1 = np.arange(10, 100) # Array of numbers from 10 up to and excluding 100
print(arr1)
```

```python
# Creating a 5x5 array of zeroes
arr2 = np.zeros((5,5))
print(arr2)
```

```python
# Converting a linear array of 8 elements into a 2x2x2 3D array
arr3 = np.zeros(8) # Flat array of eight zeroes
arr3d = arr3.reshape((2,2,2)) # Restructured array
print(arr3d)
```

```python
# NumPy array indexing is identical to Python's indexing scheme
arr4 = np.arange(2, 20)
element = arr4[6]
print(element)
```

```python
# Python's concept of lists slicing is extended to NumPy.
# The slice object is constructed by providing start, stop, and step parameters to slice()
arr5 = np.arange(20)
arr_slice = slice(1, 10, 2) # Start, stop & step
element2 = arr5[6]
print(arr5[arr_slice])
```

```python
# Extracting specific rows and columns using Slicing
d = np.array([[1,2,3], [3,4,5], [4,5,6]])
print(d[0:2, 0:2]) # Slice the first two rows and the first two columns
```

```python
print(d.shape) # Returns a tuple consisting of array dimensions
print(d.ndim) # Attribute returns the number of array dimensions
print(d.itemsize) # Returns the length of each element of array in bytes
```

## Pandas
Pandas provides three data structure - all of which are build on top of the NumPy array - all the data structures are value-mutable

1. Series (1D) - labeled, homogenous array of immutable size
2. DataFrames (2D) - labeled, heterogeneously typed, size-mutable tabular data structures

### Series
1. A Series is a single-dimensional array structures that stores homogenous data i.e., data of a single type.
2. All the elements of a Series are value-mutable and size-immutable
3. Data can be of multiple data types such as ndarray, lists, constants, series, dict etc.
4. Indexes must be unique, hashable and have the same length as data.  Defaults to `np.arrange(n)` if no index is passed.
5. Data type of each column; if none is mentioned, it will be inferred; automatically.
6. Deep copies data, set to false as default

```python
# Creating an empty Series
series = pd.Series() # The Series() function creates a new Series
print(series)
```

```python
# Creating a series from an ndarray
# Note that indexes are a assigned automatically if not specifies
arr = np.array([10,20,30,40,50])
series1 = pd.Series(arr)
print(series1)
```

```python
# Creating a series from a Python dict
# Note that the keys of the dictionary are used to assign indexes during conversion
data = {'a':10, 'b':20, 'c':30}
series2 = pd.Series(data)
print(series2)
```

## DataFrames
1. A DataFrame is a 2D data structure in which data is aligned in a tabular fashion consisting of rows & columns.
2. A DataFrame can be created using the following constructor - pandas.DataFrame(data, index, dtype, copy).
3. Data can be of multiple data types such as ndarray, list, constants, series, dict etc.
4. Index Row and column labels of the dataframe; defaults to np.arrange(n) if no index is passed
5. Data type of each column.
6. Creates a deep copy of the data, set to false as default.

```python
# Converting a list into a DataFrame
list1 = [10, 20, 30, 40]
table = pd.DataFrame(list1)
print(table)
```

```python
# Creating a DataFrame from a list of dictionaries
data = [{'a':1, 'b':2}, {'a':2, 'b':4, 'c':8}]
table1 = pd.DataFrame(data)
print(table1)
# NaN (not a number) is stored in areas where no data is provided
```

```python
# Creating a DataFrame from a list of dictionaries and accompaying row indices
table2 = pd.DataFrame(data, index = ['first', 'second'])
# Dict keys become column lables
print(table2)
```

```python
# Converting a dictionary of series into a DataFrame
data1 = {'one':pd.Series([1,2,3], index = ['a', 'b', 'c']),
        'two':pd.Series([1,2,3,4], index = ['a', 'b', 'c', 'd'])}
table3 = pd.DataFrame(data1)
print(table3)
# the resultant index is the union of all the series indexes passed
```

```python
# A new column can be added to a DataFrame when the data is passed as a Series
table3['three'] = pd.Series([10,20,30, 40], index = ['a', 'b', 'c', 'd'])
print(table3)
```

```python
# DataFrame columns can be deleted using the del() function
del table3['one']
print(table3)
```

```python
# DataFrame columns can be deleted using the pop() function
table3.pop('two')
print(table3)
```
